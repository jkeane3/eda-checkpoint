```python
# Check for null values
df.isnull().sum().sum()
```

```python
# Get the percentage of missing values for each column
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Look for outliers in population
plt.figure(figsize = (20,10))
p = df['TotalPopulation'].hist(bins=100)
```

```python
# Plot a heatmap of any missing data to see it visually
# Previous EDA shows low amount of missing data
cols = df.columns[:30] #first 30 columns
colors = ['#000099','#ffff00'] #yellow is missing, blue is not missing
sns.heatmap(df[cols].isnull(), cmap=sns.color_palette(colors))
```
